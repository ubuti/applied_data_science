```python
import numpy as np
from random import choice
from sklearn.datasets import load_sample_image
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from PIL import Image
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

def load_image(image_path):
    if not image_path:
        return np.array(Image.fromarray(load_sample_image(choice(('china.jpg', 'flower.jpg')))))
    img = Image.open(image_path)
    return np.array(img)

def flatten_image(image):
    h, w, c = image.shape
    return image.reshape(h * w, c)

def apply_pca(image_flattened, n_components):
    scaler = StandardScaler()
    image_scaled = scaler.fit_transform(image_flattened)
    pca = PCA(n_components=n_components)
    reduced_data = pca.fit_transform(image_scaled)
    return reduced_data, pca

def reconstruct_image(reduced_data, pca, scaler, original_shape):
    reconstructed_scaled = pca.inverse_transform(reduced_data)
    reconstructed = scaler.inverse_transform(reconstructed_scaled)
    return reconstructed.reshape(original_shape)

def display_images(original, reconstructed, n_components):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(original.astype(np.uint8))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title(f"Reconstructed (n_components={n_components})")
    plt.imshow(reconstructed.astype(np.uint8))
    plt.axis('off')
    plt.show()

image = load_image(None) #load_image("/Users/inuit/Pictures/Nikon Transfer 2/003/DSC_2620.JPG")
original_shape = image.shape
image_flattened = flatten_image(image)

# Precompute reconstructions for n_components 1, 2, 3
reconstructed_images = {}
scaler = StandardScaler().fit(image_flattened)
for n in range(1, 4):
    reduced_data, pca = apply_pca(image_flattened, n)
    reconstructed = reconstruct_image(reduced_data, pca, scaler, original_shape)
    reconstructed_images[n] = reconstructed

# Interactive slider
def update(n_components=1):
    display_images(image, reconstructed_images[n_components], n_components)

interact(update, n_components=IntSlider(min=1, max=3, step=1, value=1))
```
